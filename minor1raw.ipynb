```python
# use bokeh

import json
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

class Data(object):  # each data object contains a time and distance
    
    def __init__(self):
        self.time = 0
        self.distance = 0

beaconNum = "0:0" 
allBeacons = {}  # dictionary of every beacon in a block

def plotBlock():
    global allBeacons
    global beaconNum
    x = []
    y = []
    tempTitle="Beacon " + str(beaconNum)
    output_notebook()
    p = figure(
        tools="pan, box_zoom, wheel_zoom, reset, resize",
        title=tempTitle, x_axis_label='Time (seconds)', y_axis_label='Distance'
    )
    for i in range(len(allBeacons[beaconNum])-1):  # position in allBeacon[beaconNum] array
        x.append(allBeacons[beaconNum][i].time)
        y.append(allBeacons[beaconNum][i].distance)
    p.scatter(x, y, marker="circle", fill_color="blue", line_width=1)
    show(p)

def main():
    global beaconNum
    global allBeacons

    f = open("minor1.txt")
    first = True
    
    for line in f:
        words = line.split()
        time = int(words[0])  # get time
        if first == True:
            initialTime = time  # set base time
            first = False
        time = time - initialTime  # get elapsed time in ms
        time = time / 1000 # elapsed time in seconds
        json_string = words[1]  # get json text
        parsed_json = json.loads(json_string)  # parse json
        beaconNum = str(parsed_json['major']) + ":" + str(parsed_json['minor'])  # beacon identified by major and minor
        rssi = parsed_json['rssi']
        distanceRatio = 10**((rssi-60)/-20)  # reverse of log-distance pathloss model
        allBeacons.setdefault(beaconNum, []).append(Data())  # each beacon has an array of Data
        dataPos = len(allBeacons[beaconNum]) - 1  # position in array
        allBeacons[beaconNum][dataPos].time = time
        allBeacons[beaconNum][dataPos].distance = distanceRatio
    
    plotBlock()
    f.close()  # close file
        
if __name__ == "__main__":
    main()
```
